```python
import numpy as np , matplotlib.pyplot as plt , pandas as pd
from IPython.display import display, HTML
```

```python
def a_statistics(data):
  stats={
      "Min":np.min(data),
      "Max":np.max(data),
      "q1":np.percentile(data,25),
      "q3":np.percentile(data,75),
      "Range":np.max(data)-np.min(data),
      "Variance":np.var(data)
  }
  return stats
```

```python
a_summary_data=pd.read_csv('Cumulative_WHO_data.csv')
```

```python
a_total_cases=a_summary_data['Cases - cumulative total']
a_total_cases_bins=[i for i in range(0,100000000,2000000)]
plt.hist(a_total_cases,bins=a_total_cases_bins,edgecolor='black')
plt.xlabel('total number of cases in a country')
plt.ylabel('number of countries')
```

```python
a_total_cases_CLT_size=70
a_total_cases_bins1=[i for i in range(0,30000000,1000000)]
a_random_selection_total_cases=[np.mean(np.random.choice(a_total_cases,size=a_total_cases_CLT_size,replace=True)) for i in range(0,5000)]
plt.hist(a_random_selection_total_cases,bins=a_total_cases_bins1,edgecolor='black')
plt.xlabel(f'The average of {a_total_cases_CLT_size} countrys\' total cases')
plt.ylabel('frequency')
```

```python
a_total_deaths=a_summary_data['Deaths - cumulative total']
a_total_deaths_bins=[i for i in range(0,500000,10000)]
plt.hist(a_total_deaths,bins=a_total_deaths_bins,edgecolor='black')
```

```python
a_total_deaths_CLT_size=70
a_total_deaths_bins1=[i for i in range(0,240000,6000)]
random_selection_total_deaths=[np.mean(np.random.choice(a_total_deaths,size=a_total_deaths_CLT_size,replace=True)) for i in range(0,5000)]
plt.hist(random_selection_total_deaths,bins=a_total_deaths_bins1,edgecolor='black')
plt.xlabel(f'The average of {a_total_deaths_CLT_size} countrys\' total deaths')
plt.ylabel('frequency')
```

```python
a_cases_per_million=a_summary_data['Cases - cumulative total per 100000 population']
a_cases_per_million_bins=[i for i in range(0,100000,5000)]
plt.hist(a_cases_per_million,bins=a_cases_per_million_bins,edgecolor='black')
plt.xlabel('number of cases per million in a country')
plt.ylabel('number of countries')
```
